```python
import google.generativeai as genai
import json, os, random, time
import numpy as np

genai.configure(
    api_key=os.getenv('GOOGLE_API_KEY'))

def ask(prompt, temp=0.1):
    model = genai.GenerativeModel(
        'gemini-2.0-flash',
        generation_config={'temperature': temp})
    return model.generate_content(prompt).text.strip()

def build_few_shot(task, examples, query,
                   in_label='Input',
                   out_label='Output'):
    prompt = task + '\n\n'
    for inp, out in examples:
        prompt += f'{in_label}: {inp}\n'
        prompt += f'{out_label}: {out}\n\n'
    prompt += f'{in_label}: {query}\n{out_label}:'
    return prompt
```

```python
import tiktoken
enc = tiktoken.get_encoding('cl100k_base')

examples = {
    'English': [
        'This product is amazing! -> positive',
        'Terrible quality, want refund. -> negative',
        'It works fine for the price. -> neutral',
    ],
    'Hindi': [
        '\u092f\u0939 \u092a\u094d\u0930\u094b\u0921\u0915\u094d\u091f \u092c\u0939\u0941\u0924 \u0905\u091a\u094d\u091b\u093e \u0939\u0948! -> positive',
        '\u092c\u0939\u0941\u0924 \u0916\u0930\u093e\u092c \u0915\u094d\u0935\u093e\u0932\u093f\u091f\u0940\u0964 -> negative',
        '\u0915\u0940\u092e\u0924 \u0915\u0947 \u0939\u093f\u0938\u093e\u092c \u0938\u0947 \u0920\u0940\u0915 \u0939\u0948\u0964 -> neutral',
    ],
    'Telugu': [
        '\u0c08 \u0c09\u0c24\u0c4d\u0c2a\u0c24\u0c4d\u0c24\u0c3f \u0c1a\u0c3e\u0c32\u0c3e \u0c2c\u0c3e\u0c17\u0c41\u0c02\u0c26\u0c3f! -> positive',
        '\u0c1a\u0c3e\u0c32\u0c3e \u0c1a\u0c46\u0c21\u0c4d\u0c21 \u0c28\u0c3e\u0c23\u0c4d\u0c2f\u0c24\u0964 -> negative',
        '\u0c27\u0c30\u0c15\u0c41 \u0c24\u0c17\u0c3f\u0c28\u0c1f\u0c4d\u0c32\u0c41\u0c17\u0c3e \u0c09\u0c02\u0c26\u0c3f\u0964 -> neutral',
    ],
}

en_tokens = None
for lang, exs in examples.items():
    combined = '\n'.join(exs)
    tokens = len(enc.encode(combined))
    if lang == 'English':
        en_tokens = tokens
    mult = tokens / en_tokens if en_tokens else 1
    print(f'  {lang:8s}: {tokens:4d} tokens '
          f'({mult:.1f}x vs English)')

print('\nMonthly cost (100K calls):')
for lang, exs in examples.items():
    tokens = len(enc.encode('\n'.join(exs)))
    no_cache = tokens * 100_000 * 0.15 / 1e6
    cached = tokens * 100_000 * 0.0375 / 1e6
    print(f'  {lang:8s}: ${no_cache:.2f}/mo '
          f'-> ${cached:.2f} cached')
```
