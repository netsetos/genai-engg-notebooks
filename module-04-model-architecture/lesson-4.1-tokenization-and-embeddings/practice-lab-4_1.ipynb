<details><summary>Solution</summary>


```python
from collections import Counter

class FullBPE:
    SPECIAL = ["[PAD]", "[UNK]", "[BOS]", "[EOS]"]

    def __init__(self):
        self.merges = {}
        self.vocab = {}
        self.id_to_token = {}

    def _get_pairs(self, tokens):
        pairs = Counter()
        for i in range(len(tokens) - 1):
            pairs[(tokens[i], tokens[i+1])] += 1
        return pairs

    def _merge_pair(self, tokens, pair):
        merged, i = [], 0
        while i < len(tokens):
            if (i < len(tokens) - 1
                and (tokens[i], tokens[i+1]) == pair):
                merged.append(tokens[i] + tokens[i+1])
                i += 2
            else:
                merged.append(tokens[i])
                i += 1
        return merged

    def train(self, text, num_merges=15):
        tokens = list(text)
        for step in range(num_merges):
            pairs = self._get_pairs(tokens)
            if not pairs:
                break
            best = pairs.most_common(1)[0][0]
            tokens = self._merge_pair(tokens, best)
            self.merges[best] = best[0] + best[1]
            print(f"  Merge {step+1}: "
                  f"'{best[0]}'+'{best[1]}' -> "
                  f"'{best[0]+best[1]}' "
                  f"({len(tokens)} tokens)")
        all_t = self.SPECIAL + sorted(set(tokens))
        self.vocab = {t: i for i, t in enumerate(all_t)}
        self.id_to_token = {i: t for t, i in self.vocab.items()}
        print(f"\nVocab: {len(self.SPECIAL)} special "
              f"+ {len(self.vocab)-len(self.SPECIAL)} "
              f"learned = {len(self.vocab)}")
        return tokens

    def encode(self, text):
        tokens = list(text)
        for pair, merged in self.merges.items():
            tokens = self._merge_pair(tokens, pair)
        unk_id = self.vocab["[UNK]"]
        ids = [self.vocab.get(t, unk_id) for t in tokens]
        bos = self.vocab["[BOS]"]
        eos = self.vocab["[EOS]"]
        return [bos] + ids + [eos]

    def decode(self, ids):
        bos = self.vocab["[BOS]"]
        eos = self.vocab["[EOS]"]
        tokens = [
            self.id_to_token.get(i, "[UNK]")
            for i in ids if i not in (bos, eos)
        ]
        return "".join(tokens)

# Train
bpe = FullBPE()
text = "lower lowest newer newest low new"
print("Training BPE:\n")
bpe.train(text, num_merges=6)

# Test roundtrip
print("\nRoundtrip tests:")
for word in ["lower", "newest", "low"]:
    ids = bpe.encode(word)
    back = bpe.decode(ids)
    ok = "\u2705" if back == word else "\u274c"
    print(f"  '{word}' -> {ids} -> '{back}' {ok}")

# Unknown handling
print("\nUnknown handling:")
ids = bpe.encode("xyz\U0001f600")
print(f"  'xyz\U0001f600' -> {ids}")
print(f"  (UNK id = {bpe.vocab['[UNK]']})")

```

<details><summary>Solution</summary>


```python
from google import genai
import numpy as np
import os

client = genai.Client(
    api_key=os.environ["GOOGLE_API_KEY"])

sentences = [
    "Python is a programming language",
    "Java is used for enterprise software",
    "The cat sat on the mat",
    "A dog played in the park",
    "Machine learning predicts outcomes",
    "Deep learning uses neural networks",
    "Mumbai is the financial capital",
    "Delhi is the political capital",
    "Biryani is a popular rice dish",
    "Dosa is a South Indian crepe",
    "Cricket is India's favorite sport",
    "Football is popular worldwide",
    "The stock market crashed yesterday",
    "Bitcoin reached a new high",
    "Transformers changed NLP forever",
    "BERT was the first big pretrained model",
    "Hyderabad is known for IT and biryani",
    "Bangalore is India's Silicon Valley",
    "Tokenization splits text into pieces",
    "Embeddings capture word meaning",
]

def get_emb(text):
    r = client.models.embed_content(
        model="text-embedding-004",
        contents=text)
    return np.array(r.embeddings[0].values)

def cosine_matrix(vecs):
    norms = np.linalg.norm(
        vecs, axis=1, keepdims=True)
    normed = vecs / norms
    return normed @ normed.T

print("Embedding all 20 sentences...")
embs = np.array([get_emb(s) for s in sentences])
full_dim = embs.shape[1]
full_sim = cosine_matrix(embs)
print(f"Full: {full_dim} dims\n")

print(f"{'Dims':>5} {'Corr':>8} {'Storage':>10}"
      f" {'vs Full':>8}")
print("-" * 40)
for dims in [128, 256, 384, 512, 768]:
    d = min(dims, full_dim)
    trunc = embs[:, :d]
    trunc_sim = cosine_matrix(trunc)
    corr = np.corrcoef(
        full_sim.flatten(),
        trunc_sim.flatten()
    )[0, 1]
    kb = d * len(sentences) * 4 / 1024
    print(f"  {d:>4}d  {corr:.4f}  {kb:>7.1f}KB"
          f"  {corr*100:>6.1f}%")

print("\nSweet spot: 384-512 dims")
print("  97-99% accuracy, 50-67% storage savings")
print("  Use for Module 8 RAG pipeline")
```
