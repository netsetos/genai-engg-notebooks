<details><summary>Solution</summary>


```python
np.random.seed(42)
real_mean, real_std = 5.0, 1.0
gen_mean, gen_std = 0.0, 3.0
lr = 0.1

history = {"epoch": [], "gen_mean": [], "gen_std": []}

for epoch in range(50):
    fake = np.random.normal(gen_mean, abs(gen_std), 200)
    real = np.random.normal(real_mean, real_std, 200)
    mean_diff = np.mean(real) - np.mean(fake)
    std_diff = np.std(real) - np.std(fake)
    gen_mean += lr * mean_diff
    gen_std += lr * std_diff
    history["epoch"].append(epoch)
    history["gen_mean"].append(gen_mean)
    history["gen_std"].append(abs(gen_std))

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(history["epoch"], history["gen_mean"],
             'b-', linewidth=2, label='Generator')
axes[0].axhline(y=real_mean, color='r',
                linestyle='--', label=f'Target={real_mean}')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Mean')
axes[0].set_title('Mean Convergence')
axes[0].legend()

axes[1].plot(history["epoch"], history["gen_std"],
             'g-', linewidth=2, label='Generator')
axes[1].axhline(y=real_std, color='r',
                linestyle='--', label=f'Target={real_std}')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Std')
axes[1].set_title('Std Convergence')
axes[1].legend()

final_fake = np.random.normal(gen_mean, abs(gen_std), 5000)
final_real = np.random.normal(real_mean, real_std, 5000)
axes[2].hist(final_real, bins=50, alpha=0.5,
             label='Real', color='blue')
axes[2].hist(final_fake, bins=50, alpha=0.5,
             label='Generated', color='orange')
axes[2].set_title('Final Distribution Overlap')
axes[2].legend()

plt.tight_layout()
plt.savefig('gan_convergence.png', dpi=150)
plt.show()

# Convergence summary
mean_err = abs(gen_mean - real_mean)
std_err = abs(abs(gen_std) - real_std)
converged_at = next(
    (i for i, m in enumerate(history["gen_mean"])
     if abs(m - real_mean) / real_mean < 0.05),
    50
)
print(f"\nConvergence Summary:")
print(f"  Start: mean_err=5.00, std_err=2.00")
print(f"  Final: mean_err={mean_err:.3f},"
      f" std_err={std_err:.3f}")
print(f"  Converged (<5% error) at epoch {converged_at}")
```

<details><summary>Solution</summary>


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

transform = transforms.ToTensor()
train_data = datasets.MNIST(
    './data', train=True, download=True,
    transform=transform
)
loader = DataLoader(train_data, batch_size=128,
                    shuffle=True)

class VAE(nn.Module):
    def __init__(self, latent_dim=2):
        super().__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc_mu = nn.Linear(256, latent_dim)
        self.fc_logvar = nn.Linear(256, latent_dim)
        self.fc3 = nn.Linear(latent_dim, 256)
        self.fc4 = nn.Linear(256, 784)

    def encode(self, x):
        h = F.relu(self.fc1(x))
        return self.fc_mu(h), self.fc_logvar(h)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h = F.relu(self.fc3(z))
        return torch.sigmoid(self.fc4(h))

    def forward(self, x):
        mu, logvar = self.encode(x.view(-1, 784))
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

def vae_loss(recon_x, x, mu, logvar):
    bce = F.binary_cross_entropy(
        recon_x, x.view(-1, 784), reduction='sum'
    )
    kld = -0.5 * torch.sum(
        1 + logvar - mu.pow(2) - logvar.exp()
    )
    return bce + kld

model = VAE(latent_dim=2)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(10):
    total_loss = 0
    for batch, _ in loader:
        optimizer.zero_grad()
        recon, mu, logvar = model(batch)
        loss = vae_loss(recon, batch, mu, logvar)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg = total_loss / len(train_data)
    print(f"Epoch {epoch+1:2d}: loss={avg:.2f}")

# Generate 16 new digits
with torch.no_grad():
    z = torch.randn(16, 2)
    samples = model.decode(z).view(16, 28, 28)

fig, axes = plt.subplots(4, 4, figsize=(6, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(samples[i].numpy(), cmap='gray')
    ax.axis('off')
plt.suptitle('VAE Generated Digits')
plt.tight_layout()
plt.savefig('vae_generated.png', dpi=150)
plt.show()

# Interpolate between two latent points
with torch.no_grad():
    z1, z2 = torch.randn(1, 2), torch.randn(1, 2)
    fig, axes = plt.subplots(1, 10, figsize=(12, 1.5))
    for i, t in enumerate(np.linspace(0, 1, 10)):
        z = z1 * (1 - t) + z2 * t
        img = model.decode(z).view(28, 28)
        axes[i].imshow(img.numpy(), cmap='gray')
        axes[i].axis('off')
        axes[i].set_title(f't={t:.1f}', fontsize=8)
    plt.suptitle('Latent Space Walk')
    plt.tight_layout()
    plt.savefig('vae_interpolation.png', dpi=150)
    plt.show()
print("\nVAE complete! Generated digits + interpolation.")
```
