## 1 - Measure the governance gap

Before any framework, ask the one question that cannot be faked: what share of your AI systems has actually been reviewed? Compliance is a property of *one* system; governance is a property of the whole *portfolio*. This cell computes portfolio coverage and names the ungoverned systems as "shadow AI."

```python
# THE GOVERNANCE GAP: this module gave you techniques (bias, explainability, privacy, provenance) and the law (the EU AI Act).
# Governance is what makes them REPEATABLE across a whole portfolio - not one careful engineer remembering, but a program.
portfolio = [  # (AI system, has a completed governance review?)
    ("resume screener", True), ("support chatbot", True), ("fraud model", True), ("pricing model", True),
    ("marketing copy generator", False), ("internal RAG assistant", False), ("lead-scoring model", False),
    ("demand forecaster", False), ("churn predictor", False), ("doc summarizer", False),
    ("voice IVR bot", False), ("recommendation engine", False)]
reviewed = sum(1 for _, ok in portfolio if ok)
total = len(portfolio)
print("AI systems in the organization: {}".format(total))
print("  with a completed governance review: {}".format(reviewed))
print("  ungoverned (shadow AI): {}".format(total - reviewed))
print("  governance coverage: {:.0%}".format(reviewed / total))
print()
print("Compliance is a property of one system; GOVERNANCE is a property of the PORTFOLIO. {} of {} systems have no review -".format(total - reviewed, total))
print("that is the governance gap. A program turns 'someone remembered to check' into 'every system is covered, by design'.")

# Output:
# AI systems in the organization: 12
#   with a completed governance review: 4
#   ungoverned (shadow AI): 8
#   governance coverage: 33%
#
# Compliance is a property of one system; GOVERNANCE is a property of the PORTFOLIO. 8 of 12 systems have no review -
# that is the governance gap. A program turns 'someone remembered to check' into 'every system is covered, by design'.
```

**Explanation**

This is a coverage measurement over a list of systems, not a model of any one system. It reframes the whole lesson: a carefully governed flagship model tells you nothing about the portfolio, and coverage is the honest single number that only moves when something systematic reviews every system.

**How the code works - step by step**
- **`portfolio`** - a list of 12 `(system name, reviewed?)` tuples; the first four are `True`, the rest `False`.
- **`reviewed = sum(...)`** - counts how many tuples have a completed review.
- **`total`** - the length of the portfolio.
- **`total - reviewed`** - the shadow-AI count: systems running with no review.
- **the prints** - report total, reviewed, ungoverned, and `reviewed/total` as a percentage, then state the moral: compliance is per-system, governance is per-portfolio.

**In one line:** count the reviewed systems, divide by the total, and the gap is everything left over.

**What you'll see:** Twelve systems, 4 reviewed, 8 ungoverned, and governance coverage of 33% - followed by the line that compliance is per-system while governance is per-portfolio.

## 2 - Structure the work with the NIST AI RMF

Once you accept governance must cover the whole portfolio, you need a structure that says *what the work is*. The NIST AI Risk Management Framework organizes it into four functions - GOVERN, MAP, MEASURE, MANAGE - and this cell drops eight concrete activities into them, showing GOVERN is the cross-cutting foundation, not a stage you finish.

```python
# THE NIST AI RISK MANAGEMENT FRAMEWORK: the reference (voluntary) framework organizes AI governance into FOUR functions -
# GOVERN (culture, policy, roles - the foundation), MAP (context and risks), MEASURE (assess and track), MANAGE (act on risks).
FUNCTIONS = {"GOVERN": "culture, policy, roles, accountability (the cross-cutting foundation)",
             "MAP":    "identify context, intended use, and the risks",
             "MEASURE":"assess, benchmark, and monitor the risks",
             "MANAGE": "prioritize, treat, and respond to the risks"}
activities = [  # each governance activity maps to one NIST function
    ("write the responsible-AI policy",      "GOVERN"), ("assign a model owner",          "GOVERN"),
    ("document intended use + context",      "MAP"),    ("identify affected groups",       "MAP"),
    ("run the bias + robustness evals",      "MEASURE"),("monitor drift in production",    "MEASURE"),
    ("triage and mitigate open risks",       "MANAGE"), ("run the incident-response plan", "MANAGE")]
print("The NIST AI RMF - four core functions:")
for fn, desc in FUNCTIONS.items():
    print("  {:<8} {}".format(fn, desc))
print()
print("Mapping eight activities onto the functions:")
covered = {}
for act, fn in activities:
    covered.setdefault(fn, []).append(act)
for fn in FUNCTIONS:
    print("  {:<8} {}".format(fn, ", ".join(covered.get(fn, []))))
print("All four functions are exercised. GOVERN is cross-cutting - the culture and roles that make the other three repeatable.")

# Output:
# The NIST AI RMF - four core functions:
#   GOVERN   culture, policy, roles, accountability (the cross-cutting foundation)
#   MAP      identify context, intended use, and the risks
#   MEASURE  assess, benchmark, and monitor the risks
#   MANAGE   prioritize, treat, and respond to the risks
#
# Mapping eight activities onto the functions:
#   GOVERN   write the responsible-AI policy, assign a model owner
#   MAP      document intended use + context, identify affected groups
#   MEASURE  run the bias + robustness evals, monitor drift in production
#   MANAGE   triage and mitigate open risks, run the incident-response plan
# All four functions are exercised. GOVERN is cross-cutting - the culture and roles that make the other three repeatable.
```

**Explanation**

This is a bucketing exercise: it maps a flat list of governance activities onto the four NIST functions to show the framework is exhaustive and repeatable. The key idea it encodes is that MEASURE and MANAGE are largely the LLMOps machinery from Module 14 - governance assigns and repeats your engineering rather than replacing it.

**How the code works - step by step**
- **`FUNCTIONS`** - a dict of the four NIST functions to their one-line descriptions; GOVERN is labelled the cross-cutting foundation.
- **`activities`** - eight `(activity, function)` tuples, two per function (policy/owner -> GOVERN, context/affected-groups -> MAP, evals/drift -> MEASURE, triage/incident -> MANAGE).
- **first loop** - prints each function and its description.
- **`covered.setdefault(...)`** - groups the activities under their function key.
- **second loop** - prints, per function, the activities that landed in it.
- **closing print** - notes all four functions are exercised and GOVERN is what makes the others repeatable.

**In one line:** list the four functions, then sort eight activities into them to see the framework is complete.

**What you'll see:** The four NIST functions with their descriptions, then each function listed with its two mapped activities, and a closing line that GOVERN is the cross-cutting foundation.

## 3 - Score a model card's completeness

The framework says "document your systems"; the model card is the artifact that does it. This cell scores one card against the eight sections a governance program requires - and the revealing part is *which* two sections are missing.

```python
# MODEL CARDS: the core transparency artifact. A model card documents what a model is FOR, what it was trained on, how it was
# evaluated, and where it must NOT be used. Score a card against the sections a governance program requires - gaps are risks.
REQUIRED = ["intended use + users", "training data + provenance", "evaluation metrics", "fairness / bias assessment",
            "limitations + out-of-scope uses", "ethical + safety considerations", "owner + contact", "version + date"]
present = {"intended use + users", "training data + provenance", "evaluation metrics", "fairness / bias assessment",
           "owner + contact", "version + date"}                 # this card is missing two sections
missing = [s for s in REQUIRED if s not in present]
print("Model-card completeness check ({} required sections):".format(len(REQUIRED)))
for s in REQUIRED:
    print("  [{}] {}".format("x" if s in present else " ", s))
score = len(present) / len(REQUIRED)
print()
print("completeness: {}/{} = {:.0%}  ->  {}".format(len(present), len(REQUIRED), score, "COMPLETE" if not missing else "INCOMPLETE, {} gap(s)".format(len(missing))))
for m in missing:
    print("   - missing: " + m)
print("A model without 'limitations' and 'ethical considerations' looks documented but hides exactly where it fails - fill the gaps.")

# Output:
# Model-card completeness check (8 required sections):
#   [x] intended use + users
#   [x] training data + provenance
#   [x] evaluation metrics
#   [x] fairness / bias assessment
#   [ ] limitations + out-of-scope uses
#   [ ] ethical + safety considerations
#   [x] owner + contact
#   [x] version + date
#
# completeness: 6/8 = 75%  ->  INCOMPLETE, 2 gap(s)
#    - missing: limitations + out-of-scope uses
#    - missing: ethical + safety considerations
# A model without 'limitations' and 'ethical considerations' looks documented but hides exactly where it fails - fill the gaps.
```

**Explanation**

This is a checklist scorer over a fixed set of required sections. Its point is the failure mode: teams fill in the flattering sections and skip "limitations" and "ethical considerations" - exactly where a model quietly fails - so a completeness score turns an invisible omission into a tracked gap.

**How the code works - step by step**
- **`REQUIRED`** - the eight sections a complete model card must have.
- **`present`** - a set of the six sections this card actually filled (limitations and ethical/safety are absent).
- **`missing`** - the required sections not in `present`.
- **the loop** - prints each required section with `[x]` if present or `[ ]` if missing.
- **`score`** - `len(present)/len(REQUIRED)`, printed as a percentage with COMPLETE / INCOMPLETE.
- **final loop + print** - lists each missing section and warns a card without limitations and ethics looks documented but hides where it fails.

**In one line:** check the card against eight required sections and score the fraction present.

**What you'll see:** A checkbox list of eight sections with limitations and ethical/safety unchecked, a completeness of 6/8 = 75% marked INCOMPLETE, and the two missing sections named.

## 4 - Track risks in a register

A model card documents one system; the risk register tracks risk across all of them. Each entry gets a likelihood x impact score, an owner, and a status - and governance's job is to drive the *open high* risks to closure. This cell builds the register and counts what still needs attention.

```python
# THE AI RISK REGISTER: a living list of the organization's AI risks, each scored by LIKELIHOOD x IMPACT, with an OWNER and a
# STATUS. Governance tracks the open high risks to closure - an untracked risk is an unmanaged one. Count what needs attention.
RISKS = [  # (risk, likelihood 1-5, impact 1-5, owner, status)
    ("resume screener disparate impact", 4, 5, "ML lead",       "open"),
    ("chatbot leaks PII in logs",        3, 5, "platform",      "open"),
    ("model drift unnoticed",            4, 3, "MLOps",         "mitigating"),
    ("vendor model deprecated",          2, 3, "procurement",   "accepted"),
    ("training data licence unclear",    3, 4, "legal",         "open")]
THRESHOLD = 12                                                   # score >= 12 is a HIGH risk needing active treatment
print("AI risk register (score = likelihood x impact, high >= {}):".format(THRESHOLD))
open_high = 0
for risk, L, I, owner, status in sorted(RISKS, key=lambda r: r[1] * r[2], reverse=True):
    score = L * I
    tag = "HIGH" if score >= THRESHOLD else "moderate"
    if score >= THRESHOLD and status == "open": open_high += 1
    print("  {:<34} L{} x I{} = {:>2} {:<9} owner={:<12} {}".format(risk, L, I, score, tag, owner, status))
print()
print("{} HIGH risk(s) are still OPEN with no active treatment - the register makes that visible and assigns an owner to each.".format(open_high))
print("A risk you have not written down is a risk nobody owns. The register turns vague worry into tracked, owned, closable work.")

# Output:
# AI risk register (score = likelihood x impact, high >= 12):
#   resume screener disparate impact   L4 x I5 = 20 HIGH      owner=ML lead      open
#   chatbot leaks PII in logs          L3 x I5 = 15 HIGH      owner=platform     open
#   model drift unnoticed              L4 x I3 = 12 HIGH      owner=MLOps        mitigating
#   training data licence unclear      L3 x I4 = 12 HIGH      owner=legal        open
#   vendor model deprecated            L2 x I3 =  6 moderate  owner=procurement  accepted
#
# 3 HIGH risk(s) are still OPEN with no active treatment - the register makes that visible and assigns an owner to each.
# A risk you have not written down is a risk nobody owns. The register turns vague worry into tracked, owned, closable work.
```

**Explanation**

This is a scoring-and-filtering pass over a list of risks, sorted by severity. The real lesson it encodes is durability: writing a risk down is what assigns it - a worry aired in a meeting and never recorded evaporates the moment the room clears.

**How the code works - step by step**
- **`RISKS`** - five `(risk, likelihood, impact, owner, status)` tuples, likelihood and impact each 1-5.
- **`THRESHOLD = 12`** - a score at or above this is HIGH and needs active treatment.
- **the sorted loop** - orders risks by `likelihood * impact` descending, tags each HIGH or moderate, and prints the row with its owner and status.
- **`open_high`** - increments for each risk that is both HIGH and still `open` (drift is `mitigating`, so it does not count).
- **closing prints** - report the open-high count and the moral: an untracked risk is one nobody owns.

**In one line:** score each risk by likelihood x impact, sort, and count the high ones still open.

**What you'll see:** Five risks sorted by score (20, 15, 12, 12, 6), and a line reporting 3 HIGH risks still OPEN with no active treatment.

## 5 - Separate the three lines of defense

A register assigns owners, but ownership alone is not accountability - you also need *independence*. The three-lines model splits it: line 1 owns the risk (builders), line 2 sets policy and oversees, line 3 gives independent assurance. This cell checks each control for coverage gaps.

```python
# THE THREE LINES OF DEFENSE: accountability needs SEPARATION. Line 1 OWNS and manages the risk (the builders); line 2 sets
# policy and OVERSEES (a responsible-AI / risk function); line 3 provides INDEPENDENT assurance (internal audit). Check coverage.
CONTROLS = [  # (control, lines engaged, high_stakes -> needs independent line-3 assurance)
    ("bias evaluation",     {"L1", "L2"},       True),
    ("privacy / PII review", {"L1", "L2"},       False),
    ("incident response",   {"L1"},             True),   # GAP - no independent oversight
    ("model documentation", {"L1", "L2", "L3"}, False),
    ("regulatory mapping",  {"L1", "L2"},       False)]
def gaps(controls):
    out = []
    for name, lines, high_stakes in controls:
        if "L2" not in lines: out.append((name, "no line-2 oversight"))
        if high_stakes and "L3" not in lines: out.append((name, "no line-3 assurance on a high-stakes control"))
    return out
print("Three-lines-of-defense coverage (L1 owns, L2 oversees, L3 independently assures):")
for name, lines, _ in CONTROLS:
    print("  {:<22} {}".format(name, " ".join(sorted(lines))))
issues = gaps(CONTROLS)
print()
print("coverage gaps: {}".format("none" if not issues else "{} found:".format(len(issues))))
for name, why in issues:
    print("   - {}: {}".format(name, why))
print("The point is independence: the team that builds a model should not be the only one judging its safety. Separate the lines.")

# Output:
# Three-lines-of-defense coverage (L1 owns, L2 oversees, L3 independently assures):
#   bias evaluation        L1 L2
#   privacy / PII review   L1 L2
#   incident response      L1
#   model documentation    L1 L2 L3
#   regulatory mapping     L1 L2
#
# coverage gaps: 3 found:
#    - bias evaluation: no line-3 assurance on a high-stakes control
#    - incident response: no line-2 oversight
#    - incident response: no line-3 assurance on a high-stakes control
# The point is independence: the team that builds a model should not be the only one judging its safety. Separate the lines.
```

**Explanation**

This is a gap-finder over a list of controls, flagging any that lack oversight or independent assurance. It encodes the old audit principle that the people who build a thing should not be the only ones certifying it is safe - a control missing a line is where it can quietly rot.

**How the code works - step by step**
- **`CONTROLS`** - five `(control, set-of-lines-engaged, high_stakes?)` tuples; incident response has only `{L1}`.
- **`gaps(controls)`** - for each control, flags "no line-2 oversight" if L2 is absent, and "no line-3 assurance" if it is high-stakes but lacks L3.
- **the loop** - prints each control with the lines it engages.
- **`issues = gaps(CONTROLS)`** - collects every flagged gap.
- **closing loop + print** - lists each gap with its reason and states the point: separate the lines because builders cannot be their own auditors.

**In one line:** for each control, check it has oversight (L2) and, if high-stakes, independent assurance (L3).

**What you'll see:** A controls-to-lines listing, then 3 gaps: no line-3 assurance on bias evaluation, and no line-2 oversight plus no line-3 assurance on incident response.

## 6 - Score governance maturity by its weakest link

A program should answer "how good is our governance?" with more than a feeling. This cell scores five dimensions on a 1-5 ladder and combines them with the crucial rule: maturity is a *floor, not an average* - the weakest dimension sets the real ceiling.

```python
# GOVERNANCE MATURITY: you cannot improve what you do not measure, including governance itself. Score five dimensions on a
# 1-5 ladder (1 ad-hoc, 2 repeatable, 3 defined, 4 managed, 5 optimizing) and read the overall maturity level.
LEVELS = {1: "ad-hoc", 2: "repeatable", 3: "defined", 4: "managed", 5: "optimizing"}
dimensions = {"policy + standards": 3, "roles + accountability": 2, "tooling + automation": 4,
              "monitoring + metrics": 2, "culture + training": 3}
avg = sum(dimensions.values()) / len(dimensions)
level = min(dimensions.values())                                 # floor to the WEAKEST dimension, not the mean
print("AI governance maturity (each dimension on the 1-5 ladder):")
for dim, score in dimensions.items():
    print("  {:<24} {}  ({})".format(dim, score, LEVELS[score]))
print()
print("overall = mean {:.1f} -> level {} ({}); the weakest dimensions set the real ceiling.".format(avg, level, LEVELS[level]))
weakest = min(dimensions, key=dimensions.get)
print("Lowest: '{}' at {}. Maturity is a floor, not an average - a strong tooling score does not cover weak accountability.".format(weakest, dimensions[weakest]))
print("Assess yearly and target the weakest dimension; governance improves the same way software does - deliberately, level by level.")

# Output:
# AI governance maturity (each dimension on the 1-5 ladder):
#   policy + standards       3  (defined)
#   roles + accountability   2  (repeatable)
#   tooling + automation     4  (managed)
#   monitoring + metrics     2  (repeatable)
#   culture + training       3  (defined)
#
# overall = mean 2.8 -> level 2 (repeatable); the weakest dimensions set the real ceiling.
# Lowest: 'roles + accountability' at 2. Maturity is a floor, not an average - a strong tooling score does not cover weak accountability.
# Assess yearly and target the weakest dimension; governance improves the same way software does - deliberately, level by level.
```

**Explanation**

This cell computes both a mean and a minimum over five dimension scores, then deliberately uses the minimum. The idea it encodes is that a program is only as mature as its weakest dimension, because that is where it will actually break - strong tooling does not compensate for weak accountability.

**How the code works - step by step**
- **`LEVELS`** - maps 1-5 to the ladder names (ad-hoc, repeatable, defined, managed, optimizing).
- **`dimensions`** - five dimensions with scores: policy 3, roles 2, tooling 4, monitoring 2, culture 3.
- **`avg`** - the mean (2.8), computed to be contrasted, not used as the answer.
- **`level = min(...)`** - the overall level, floored to the weakest dimension.
- **the loop** - prints each dimension with its score and ladder name.
- **`weakest = min(dimensions, key=...)`** - names the lowest dimension (roles + accountability).
- **closing prints** - contrast mean vs floor and advise assessing yearly and targeting the weakest dimension.

**In one line:** score five dimensions, then take the minimum - not the average - as your real maturity.

**What you'll see:** Five dimensions on the ladder, then "mean 2.8 -> level 2 (repeatable)" with roles + accountability named as the weakest link at 2.

## 7 - Gate every release

Everything in the lesson - and the module - converges on one moment: the decision to ship. The governance release gate is a pre-release checklist, and its defining property is that it can say no. This cell evaluates the gate and returns GO or NO-GO.

```python
# THE GOVERNANCE RELEASE GATE: the whole program converges on one decision - can this AI system ship? A governance review
# gate checks that the artifacts exist and the risks are handled, then a human signs off. One unmet item is a NO-GO.
GATE = {                                                        # the pre-release governance checklist
    "risk tier classified (EU AI Act)": True,
    "risk register updated, no open HIGH risks": False,        # UNMET - an open high risk remains
    "model card complete": True,
    "bias + safety evals passed": True,
    "monitoring + incident plan configured": True,
    "accountable owner sign-off": True}
unmet = [k for k, ok in GATE.items() if not ok]
print("GOVERNANCE RELEASE GATE (the program's single decision point):")
for k, ok in GATE.items():
    print("  [{}] {}".format("x" if ok else " ", k))
print()
print("decision: {}".format("GO - ship" if not unmet else "NO-GO - {} unmet:".format(len(unmet))))
for u in unmet:
    print("   - " + u)
print("Governance is not a document you write once; it is a LOOP - govern, map, measure, manage - that runs at every release and")
print("in production forever. The gate is where the module's techniques, the law, and the program meet: responsible AI, operationalized.")

# Output:
# GOVERNANCE RELEASE GATE (the program's single decision point):
#   [x] risk tier classified (EU AI Act)
#   [ ] risk register updated, no open HIGH risks
#   [x] model card complete
#   [x] bias + safety evals passed
#   [x] monitoring + incident plan configured
#   [x] accountable owner sign-off
#
# decision: NO-GO - 1 unmet:
#    - risk register updated, no open HIGH risks
# Governance is not a document you write once; it is a LOOP - govern, map, measure, manage - that runs at every release and
# in production forever. The gate is where the module's techniques, the law, and the program meet: responsible AI, operationalized.
```

**Explanation**

This is a final all-checks-must-pass gate over the artifacts the whole module produced. It encodes the closing idea: a gate that is always green is a rubber stamp, and governance is not a document but a govern-map-measure-manage loop that runs at every release and in production forever.

**How the code works - step by step**
- **`GATE`** - a dict of six checks to booleans (risk tier, no open high risks, model card, evals, monitoring/incident, owner sign-off); the open-high-risks check is `False`.
- **`unmet`** - collects every check that is `False`.
- **the loop** - prints each check with `[x]` or `[ ]`.
- **the decision print** - `GO - ship` only if `unmet` is empty, otherwise `NO-GO` with the count.
- **closing loop + prints** - list each unmet item and state that governance is a continuous loop where the module's techniques, the law, and the program meet.

**In one line:** every check must pass - one unmet item forces NO-GO.

**What you'll see:** A six-item checklist with the risk-register item unchecked, giving "decision: NO-GO - 1 unmet", and a closing note that governance is a loop run at every release.

Read as a sequence, the seven checks are one program: measure coverage (who is even reviewed), structure the work with NIST's four functions, document each system in a model card, track every risk to an owner in the register, separate the builders from their auditors with three lines, score maturity by its weakest link, and gate every release with a decision that can say no. None of it calls a model - it is all plain rule-checking on illustrative data, because the point is the machinery, not the math. This is the close of Module 15: the bias tests, explanations, provenance, privacy controls, and disclosure you built earlier are the checks that meet here, at the gate, in the hands of someone accountable.
