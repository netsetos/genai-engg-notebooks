```python
from dataclasses import dataclass

@dataclass
class AIP:
    org:str; ver:str; date:str; owner:str; models:list; prohibited:list; data:list; review:str; contact:str; enforce:list
    def render(self):
        print(f"AI POLICY: {self.org} v{self.ver} ({self.date})")
        print(f"Owner: {self.owner}")
        print(f"\nApproved Models ({len(self.models)}):")
        for m in self.models: print(f"  [OK] {m}")
        print(f"\nProhibited Uses ({len(self.prohibited)}):")
        for p in self.prohibited: print(f"  [X]  {p}")
        print(f"\nData Rules ({len(self.data)}):")
        for i,d in enumerate(self.data,1): print(f"  {i}.   {d}")
        print(f"\nEnforcement:")
        for e in self.enforce: print(f"  ->   {e}")
        print(f"\nReview: {self.review} | Contact: {self.contact}")
        print(f"Total items: {len(self.models)+len(self.prohibited)+len(self.data)+len(self.enforce)}")

p=AIP("Netsetos","2.0","2026-04-01","CTO",
    ["Gemini 2.0 Flash","Gemini 2.5 Pro","Llama 3.1 8B (self-hosted)","nomic-embed-text (local)"],
    ["Auto admission without human review","Student PII to cloud LLMs",
     "Impersonating real people","AI-only grading decisions","Training on student data without consent"],
    ["No PII in cloud prompts (12.4)","All calls cost-tracked (11.1)","Student data on approved models only",
     "Data audit quarterly (12.3)","Logs retained 90 days","RAG index rebuilt monthly"],
    "Quarterly","ai-safety@netsetos.com",
    ["1st: warning + training","2nd: API access revoked 30d","3rd: management escalation","Critical: immediate revocation + legal"])
p.render()
```

```python
from dataclasses import dataclass
import json

@dataclass
class MC:
    name:str; ver:str; provider:str; use:str; users:str; limits:list; biases:list
    eval:dict; ethical:list; sources:list; oversight:str; last_eval:str
    def completeness(self):
        fs=["name","ver","provider","use","users","limits","biases","eval","ethical","sources","oversight","last_eval"]
        filled=sum(1 for f in fs if getattr(self,f)); return {"f":filled,"t":len(fs),"pct":round(filled/len(fs)*100)}
    def render(self):
        print(f"MODEL CARD: {self.name} v{self.ver}")
        print(f"Provider: {self.provider}")
        print(f"Use: {self.use}")
        print(f"Users: {self.users}")
        print(f"\nEval Metrics:")
        for k,v in self.eval.items(): print(f"  {k}: {v}")
        print(f"\nLimitations ({len(self.limits)}):")
        for l in self.limits: print(f"  - {l}")
        print(f"\nKnown Biases ({len(self.biases)}):")
        for b in self.biases: print(f"  - {b}")
        print(f"\nEthical: {', '.join(self.ethical)}")
        print(f"Sources: {', '.join(self.sources)}")
        print(f"Oversight: {self.oversight}")
        c=self.completeness(); print(f"\nCompleteness: {c['f']}/{c['t']} ({c['pct']}%)")

c=MC("Netsetos Chatbot","3.1","Gemini 2.0 Flash",
    "Answer student questions about courses, pricing, refunds",
    "Prospective and enrolled students",
    ["Text-only input","May hallucinate pricing","English > Hindi quality","No real-time payment"],
    ["English responses 25 words longer (p=0.03)","No name-origin sentiment bias (p=0.42)","Lower quality for code-mixed queries"],
    {"accuracy":"92%","hallucination":"3%","latency":"1.2s","satisfaction":"4.3/5","cost":"Rs0.14/query"},
    ["Not for admission","Human escalation available","AI disclosed","90-day retention"],
    ["Course catalog","FAQ DB","Pricing API","Schedule DB"],
    "Support team reviews flagged conversations daily","2026-03-15")
c.render()
```

```python
from dataclasses import dataclass
from enum import Enum

class CS(Enum):
    PASS="pass"; PARTIAL="partial"; FAIL="fail"

@dataclass
class AC:
    cat:str; item:str; ref:str; status:CS; owner:str; notes:str=""

def run_audit():
    checks=[
        AC("Bias","Bias test suite monthly","12.1",CS.PASS,"ML Eng","Last: 2026-03-15"),
        AC("Bias","No p<0.05 bias","12.1",CS.PASS,"ML Eng","p=0.42"),
        AC("Bias","Bias in CI/CD","12.1",CS.PARTIAL,"DevOps","Not in CI yet"),
        AC("Privacy","PII detector active","12.4",CS.PASS,"ML Eng","6 types"),
        AC("Privacy","Anonymization deployed","12.4",CS.PASS,"ML Eng","Reversible"),
        AC("Privacy","No PII in logs","12.4",CS.PASS,"DevOps","Verified"),
        AC("Security","Input guardrails","9.11",CS.PASS,"Security","Injection blocked"),
        AC("Security","Output filtering","9.11",CS.PASS,"ML Eng","Copyright+PII"),
        AC("Security","Rate limiting","10.3",CS.PASS,"DevOps","Per-team RPM"),
        AC("Transparency","AI disclosed","12.6",CS.PASS,"Frontend","Banner"),
        AC("Transparency","Model card published","12.7",CS.PASS,"ML Eng","v3.1"),
        AC("Transparency","Confidence indicators","10.10",CS.PARTIAL,"Frontend","Chatbot only"),
        AC("Reliability","Circuit breaker","10.2",CS.PASS,"DevOps","3 providers"),
        AC("Reliability","Fallback providers","10.2",CS.PASS,"DevOps","Gemini->OAI->Claude"),
        AC("Reliability","Monitoring active","10.8",CS.PASS,"DevOps","5 metrics"),
        AC("Cost","Cost tracking","11.1",CS.PASS,"ML Eng","Per-request"),
        AC("Cost","Team budgets","11.1",CS.PASS,"Finance","Monthly"),
        AC("Cost","Anomaly detection","11.1",CS.FAIL,"DevOps","Not implemented"),
        AC("Legal","Data audited","12.3",CS.PASS,"Data Eng","10 sources"),
        AC("Legal","EU AI Act classified","12.6",CS.PARTIAL,"Legal","Chatbot done"),
        AC("Legal","TOS updated","12.3",CS.FAIL,"Legal","AI section missing"),
    ]
    print("Governance Audit:")
    cats={}
    for c in checks: cats.setdefault(c.cat,[]).append(c)
    icons={"pass":"[X]","partial":"[~]","fail":"[!]"}
    for cat,items in cats.items():
        cp=sum(1 for c in items if c.status==CS.PASS)
        print(f"\n  {cat} ({cp}/{len(items)}):")
        for c in items:
            print(f"    {icons[c.status.value]} {c.item}")
            if c.status!=CS.PASS: print(f"       {c.owner}: {c.notes}")

    p=sum(1 for c in checks if c.status==CS.PASS)
    pa=sum(1 for c in checks if c.status==CS.PARTIAL)
    f=sum(1 for c in checks if c.status==CS.FAIL)
    score=round((p+pa*0.5)/len(checks)*100)
    print(f"\n  Score: {score}% ({p}pass+{pa}partial+{f}fail) {'PASS' if score>=80 else 'FAIL'}")
    gaps=[c for c in checks if c.status!=CS.PASS]
    print(f"  Gaps ({len(gaps)}):")
    for c in gaps:
        pri="CRITICAL" if c.status==CS.FAIL else "MODERATE"
        print(f"    [{pri}] {c.item} -> {c.owner} (Lesson {c.ref})")

run_audit()
```

```python
from dataclasses import dataclass
from datetime import datetime

@dataclass
class GD:
    project:str
    def policy(self):
        return {"ver":"2.0","status":"ACTIVE","models":4,"prohibited":5,"data_rules":6,"review":"2026-07-01"}
    def models(self):
        return [{"name":"Chatbot","model":"Flash","card":True,"bias":True,"status":"PROD"},
            {"name":"ContentGen","model":"Pro","card":True,"bias":True,"status":"PROD"},
            {"name":"Embeddings","model":"nomic","card":True,"bias":False,"status":"PROD"},
            {"name":"Grader","model":"Pro","card":False,"bias":False,"status":"STAGING"}]
    def audit(self):
        return {"score":83,"verdict":"PASS","gaps":5,"critical":2,
            "cats":{"Bias":83,"Privacy":100,"Security":100,"Transparency":83,"Reliability":100,"Cost":83,"Legal":50}}
    def risks(self):
        return [{"risk":"Hallucination in pricing","status":"Mitigated"},
            {"risk":"PII leak to cloud","status":"Mitigated"},
            {"risk":"Bias in admission","status":"Open"},
            {"risk":"EU AI Act non-compliance","status":"In progress"},
            {"risk":"Cost overrun","status":"Partial"}]
    def compliance(self):
        return {"eu_ai_act":"LIMITED+HIGH, HIGH not compliant","gdpr":"Compliant","dpdp":"Partial"}
    def deploy_gate(self):
        a=self.audit(); m=self.models(); p=self.policy()
        checks={"audit>=80":a["score"]>=80,"no_critical":a["critical"]==0,
            "policy_active":p["status"]=="ACTIVE",
            "prod_cards":all(x["card"] for x in m if x["status"]=="PROD"),
            "prod_bias":all(x["bias"] for x in m if x["status"]=="PROD")}
        return {"checks":checks,"allowed":all(checks.values())}
    def render(self):
        print(f"GOVERNANCE DASHBOARD: {self.project}")
        print(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
        p=self.policy(); print(f"\n  Policy: v{p['ver']} ({p['status']}) Review: {p['review']}")
        print(f"\n  Models:")
        for m in self.models():
            print(f"    [{m['status']:>7}] {m['name']:<14} Card:{'Y' if m['card'] else 'N'} Bias:{'Y' if m['bias'] else 'N'}")
        a=self.audit(); print(f"\n  Audit: {a['score']}% ({a['verdict']}) Gaps:{a['gaps']}")
        for c,s in a["cats"].items(): print(f"    {c:<15} {s}% {'#'*(s//10)}")
        print(f"\n  Risks:")
        for r in self.risks(): print(f"    {'[!]' if 'Open' in r['status'] or 'progress' in r['status'] else '[X]'} {r['risk']}: {r['status']}")
        c=self.compliance(); print(f"\n  Compliance: EU={c['eu_ai_act']} | GDPR={c['gdpr']} | DPDP={c['dpdp']}")
        g=self.deploy_gate()
        print(f"\n  Deploy Gate:")
        for k,v in g["checks"].items(): print(f"    {'[X]' if v else '[!]'} {k}: {'PASS' if v else 'FAIL'}")
        print(f"  Deploy: {'YES' if g['allowed'] else 'NO'}")

d=GD("Netsetos GenAI Platform")
d.render()
```
