```python
from dataclasses import dataclass
from enum import Enum
from collections import Counter

class Risk(Enum):
    LOW="low"; MEDIUM="medium"; HIGH="high"

@dataclass
class DS:
    name:str; stype:str; license:str; risk:Risk; mitigation:str

sources=[
    DS("Python docs","Reference","PSF License",Risk.LOW,"Use freely"),
    DS("Wikipedia","Knowledge","CC BY-SA 3.0",Risk.LOW,"Attribute"),
    DS("arXiv papers","Research","Open access",Risk.LOW,"Check individual"),
    DS("Internal curriculum","Internal","Owned",Risk.LOW,"Full ownership"),
    DS("StackOverflow","Q&A","CC BY-SA 4.0",Risk.MEDIUM,"Attribute; no bulk scraping"),
    DS("GitHub repos","Code","Varies MIT/GPL",Risk.MEDIUM,"Check each license"),
    DS("Reddit","Community","API TOS",Risk.MEDIUM,"TOS restricts commercial"),
    DS("YouTube transcripts","Video","YouTube TOS",Risk.MEDIUM,"Use official API only"),
    DS("Published textbooks","Reference","Copyrighted",Risk.HIGH,"Reference only; no reproduction"),
    DS("Scraped news","Current events","Copyrighted",Risk.HIGH,"NYT precedent; license or remove"),
]

print("Data Source Audit:")
rc=Counter()
for s in sources:
    icon={"low":"G","medium":"W","high":"X"}[s.risk.value]; rc[s.risk.value]+=1
    print(f"  [{icon}] {s.name:<24} {s.risk.value:<8} {s.license}")
print(f"\n  Distribution: Low={rc['low']} Med={rc['medium']} High={rc['high']}")
high=[s for s in sources if s.risk==Risk.HIGH]
print(f"  Status: {'NEEDS ACTION' if high else 'PASS'}")
for s in high: print(f"    [HIGH] {s.name}: {s.mitigation}")
```

```python
class CF:
    def __init__(self,th=20): self.th=th; self.db={}
    def add(self,name,text):
        ws=text.lower().split(); ng=set()
        for i in range(len(ws)-self.th+1): ng.add(" ".join(ws[i:i+self.th]))
        self.db[name]=ng
    def check(self,output):
        ws=output.lower().split(); viols=[]
        for i in range(len(ws)-self.th+1):
            ng=" ".join(ws[i:i+self.th])
            for src,ngs in self.db.items():
                if ng in ngs: viols.append({"src":src,"pos":i,"text":ng[:50]}); break
        # dedup overlapping
        if viols:
            dd=[viols[0]]
            for v in viols[1:]:
                if v["pos"]>dd[-1]["pos"]+self.th//2: dd.append(v)
            viols=dd
        return {"viols":viols,"action":"BLOCK" if viols else "PASS"}

cf=CF(20)
cf.add("Textbook A","Machine learning is a subset of artificial intelligence that provides "
    "systems the ability to automatically learn and improve from experience "
    "without being explicitly programmed It focuses on the development of "
    "computer programs that can access data and use it to learn for themselves")
cf.add("Paper B","The transformer architecture relies entirely on attention mechanisms to "
    "draw global dependencies between input and output It dispenses with "
    "recurrence and convolutions entirely and instead uses multi head self "
    "attention to jointly attend to information from different representation")
cf.add("News C","The European Union has passed comprehensive artificial intelligence "
    "legislation that classifies AI systems by risk level and imposes strict "
    "requirements on high risk applications including mandatory conformity "
    "assessments bias testing and human oversight for critical decisions")

tests=[
    ("Clean","AI transforms education with personalized learning experiences"),
    ("Clean","ML lets computers learn from data without explicit rules"),
    ("VIOLATE","Machine learning is a subset of artificial intelligence that provides "
     "systems the ability to automatically learn and improve from experience "
     "without being explicitly programmed It focuses on the development of "
     "computer programs that can access data and use it to learn for themselves"),
    ("VIOLATE","The transformer architecture relies entirely on attention mechanisms to "
     "draw global dependencies between input and output It dispenses with "
     "recurrence and convolutions entirely and instead uses multi head self "
     "attention to jointly attend to information from different representation"),
    ("Clean","Python was created by Guido van Rossum and emphasizes readability"),
    ("VIOLATE","The European Union has passed comprehensive artificial intelligence "
     "legislation that classifies AI systems by risk level and imposes strict "
     "requirements on high risk applications including mandatory conformity "
     "assessments bias testing and human oversight for critical decisions"),
]

print("Copyright Filter:")
vc=0
for label,text in tests:
    r=cf.check(text)
    vc+=len(r["viols"])
    print(f"  [{r['action']:5s}] {label}: {text[:45]}...")
    for v in r["viols"]: print(f"         Source: {v['src']}")
print(f"\n  Tested: {len(tests)} | Violations: {vc} | Clean: {len(tests)-sum(1 for _,t in tests if cf.check(t)['viols'])}")
```

```python
from dataclasses import dataclass
from enum import Enum
from datetime import datetime

class St(Enum):
    DONE="done"; PENDING="pending"; BLOCKED="blocked"

@dataclass
class CI:
    id:int; cat:str; item:str; evidence:str; owner:str; status:St; notes:str=""

items=[
    CI(1,"Training Data","All sources documented","Spreadsheet","Data Eng",St.DONE,"10 sources audited"),
    CI(2,"Training Data","No pirated content","Audit report","Legal",St.DONE,"2 high-risk removed"),
    CI(3,"Training Data","robots.txt respected","Script+logs","Data Eng",St.DONE,"Crawler compliant"),
    CI(4,"Training Data","PII removed","PII scan","Privacy",St.PENDING,"Scan next week"),
    CI(5,"Generated","AI content labeled","UI screenshot","Frontend",St.DONE,"Badge on outputs"),
    CI(6,"Generated","No fake quotes","Prompt guardrails","ML Eng",St.DONE,"System prompt blocks"),
    CI(7,"Generated","Copyright filter active","Filter code","ML Eng",St.DONE,"20-word threshold"),
    CI(8,"Generated","User prompt rights","TOS clause","Legal",St.DONE,"Section 4.2"),
    CI(9,"Disclosure","AI interaction notice","UI banner","Frontend",St.DONE,"Persistent banner"),
    CI(10,"Disclosure","Model version disclosed","API header","Backend",St.DONE,"X-Model-Version"),
    CI(11,"Disclosure","Limitations documented","Help page","Writer",St.PENDING,"Draft ready"),
    CI(12,"Disclosure","Retention policy","Privacy page","Legal",St.BLOCKED,"Awaiting legal review"),
    CI(13,"Legal","TOS covers AI content","TOS document","Legal",St.DONE,"Section 3"),
    CI(14,"Legal","Indemnification clause","TOS clause","Legal",St.PENDING,"Under review"),
    CI(15,"Legal","Regulatory compliance","Compliance matrix","Compliance",St.BLOCKED,"EU AI Act mapping incomplete"),
]

print("Compliance Report:")
print(f"  Date: {datetime.now().strftime('%Y-%m-%d')}")
icons={"done":"[X]","pending":"[ ]","blocked":"[!]"}
cats={}
for i in items:
    cats.setdefault(i.cat,[]).append(i)
for cat,ci in cats.items():
    done=sum(1 for c in ci if c.status==St.DONE)
    print(f"\n  {cat} ({done}/{len(ci)})")
    for c in ci:
        print(f"  {icons[c.status.value]} {c.id:>2}. {c.item} [{c.status.value}]")

td=sum(1 for i in items if i.status==St.DONE)
bl=[i for i in items if i.status==St.BLOCKED]
pn=[i for i in items if i.status==St.PENDING]
score=td/len(items)*100
print(f"\n  Score: {td}/{len(items)} ({score:.0f}%)")
print(f"  Blocked: {len(bl)} | Pending: {len(pn)}")
ready=score>=80 and len(bl)==0
print(f"  Launch: {'READY' if ready else 'NOT READY'}")
```
