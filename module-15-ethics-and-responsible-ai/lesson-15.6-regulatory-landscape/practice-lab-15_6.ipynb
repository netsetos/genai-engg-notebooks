```python
from dataclasses import dataclass

@dataclass
class AF:
    name:str; tier:str; obs:list; deadline:str

features=[
    AF("Course chatbot","LIMITED",["Disclose AI use","Label AI responses"],"Aug 2025"),
    AF("AI tutor","MINIMAL",["No obligations"],"N/A"),
    AF("Content generator","LIMITED",["Label AI-generated","Watermark media"],"Aug 2025"),
    AF("Student assessment","HIGH",["Bias testing","Human oversight","Explainability","Appeal mechanism","Decision logging"],"Aug 2026"),
    AF("Plagiarism detector","MINIMAL",["No obligations"],"N/A"),
    AF("Career recommender","LIMITED",["Disclose AI use","Allow opt-out"],"Aug 2025"),
    AF("Admission screener","HIGH",["Conformity assessment","Bias audit","Human review rejections","Full logging","Right to explanation"],"Aug 2026"),
    AF("Spam filter","MINIMAL",["No obligations"],"N/A"),
]

print("EU AI Act Classification:")
tc={"HIGH":0,"LIMITED":0,"MINIMAL":0}; to=0
for f in features:
    tc[f.tier]+=1; to+=len(f.obs)
    icon={"HIGH":"!","LIMITED":"~","MINIMAL":"."}[f.tier]
    print(f"  [{icon}] {f.name:<22} {f.tier:<10} ({len(f.obs)} obligations) Due: {f.deadline}")

print(f"\n  Distribution: HIGH={tc['HIGH']} LIMITED={tc['LIMITED']} MINIMAL={tc['MINIMAL']}")
print(f"  By Aug 2025: {sum(1 for f in features if f.deadline=='Aug 2025')} features (LIMITED)")
print(f"  By Aug 2026: {sum(1 for f in features if f.deadline=='Aug 2026')} features (HIGH)")
print(f"  Total obligations: {to}")
```

```python
from dataclasses import dataclass
from enum import Enum

class St(Enum):
    DONE="done"; PARTIAL="partial"; GAP="gap"

@dataclass
class CI:
    id:int; name:str; reg:str; code:str; status:St; priority:str; action:str

items=[
    CI(1,"Bias testing","EU AI Act","12.1",St.DONE,"HIGH","Monthly automated tests"),
    CI(2,"Explainability","EU AI Act+GDPR","12.2",St.DONE,"HIGH","SHAP on all predictions"),
    CI(3,"Content labeling","EU AI Act","12.5",St.DONE,"MED","AI badge on outputs"),
    CI(4,"Watermarking","EU AI Act+US EO","12.5",St.PARTIAL,"MED","Text done, image pending"),
    CI(5,"PII protection","GDPR+DPDP","12.4",St.DONE,"HIGH","Anonymize all inputs"),
    CI(6,"Data audit","EU AI Act","12.3",St.DONE,"MED","10 sources audited"),
    CI(7,"Copyright filter","IP Law","12.3",St.DONE,"MED","20-word N-gram filter"),
    CI(8,"Monitoring","Google SAIF","10.8",St.PARTIAL,"HIGH","Alerting incomplete"),
    CI(9,"Red-teaming","SAIF+US EO","9.11",St.GAP,"HIGH","Not started"),
    CI(10,"Incident response","EU AI Act","10.9",St.PARTIAL,"HIGH","Playbooks untested"),
    CI(11,"Human oversight","EU AI Act","9.6",St.GAP,"HIGH","No HITL for high-risk"),
]

print("Compliance Audit:")
done=sum(1 for i in items if i.status==St.DONE)
partial=sum(1 for i in items if i.status==St.PARTIAL)
gaps=sum(1 for i in items if i.status==St.GAP)
score=round((done+partial*0.5)/len(items)*100)

icons={"done":"[X]","partial":"[~]","gap":"[!]"}
for i in items:
    print(f"  {icons[i.status.value]} {i.id:>2}. {i.name:<20} {i.status.value:<8} {i.priority}")

print(f"\n  Score: {score}% | Done:{done} Partial:{partial} Gaps:{gaps}")
print(f"\n  Gaps (action required):")
for i in items:
    if i.status!=St.DONE:
        print(f"    #{i.id} {i.name}: {i.action}")
```

```python
from dataclasses import dataclass
from enum import Enum
from datetime import datetime

class St(Enum):
    DONE="done"; PARTIAL="partial"; GAP="gap"

@dataclass
class Req:
    id:int; name:str; reg:str; status:St; owner:str; due:str

@dataclass
class Feat:
    name:str; tier:str; obs:int; ok:bool

class CR:
    def __init__(self,proj):
        self.proj=proj; self.feats=[]; self.reqs=[]
    def build(self):
        self.feats=[Feat("Chatbot","LIMITED",2,True),Feat("AI tutor","MINIMAL",0,True),
            Feat("Content gen","LIMITED",2,True),Feat("Assessment","HIGH",5,False),
            Feat("Plagiarism","MINIMAL",0,True),Feat("Career rec","LIMITED",2,True),
            Feat("Admission","HIGH",5,False),Feat("Spam","MINIMAL",0,True)]
        self.reqs=[Req(1,"Bias testing","EU AI Act",St.DONE,"ML Eng","Done"),
            Req(2,"Explainability","EU AI Act+GDPR",St.DONE,"ML Eng","Done"),
            Req(3,"Content labeling","EU AI Act",St.DONE,"Frontend","Done"),
            Req(4,"Watermarking","EU AI Act+US EO",St.PARTIAL,"ML Eng","Q3"),
            Req(5,"PII protection","GDPR+DPDP",St.DONE,"Privacy","Done"),
            Req(6,"Data audit","EU AI Act",St.DONE,"Data Eng","Done"),
            Req(7,"Copyright filter","IP Law",St.DONE,"ML Eng","Done"),
            Req(8,"Monitoring","Google SAIF",St.PARTIAL,"DevOps","Q3"),
            Req(9,"Red-teaming","SAIF+US EO",St.GAP,"Security","Q3"),
            Req(10,"Incident response","EU AI Act",St.PARTIAL,"DevOps","Q2"),
            Req(11,"Human oversight","EU AI Act",St.GAP,"Product","Q2")]
    def report(self):
        self.build()
        done=sum(1 for r in self.reqs if r.status==St.DONE)
        partial=sum(1 for r in self.reqs if r.status==St.PARTIAL)
        gaps=sum(1 for r in self.reqs if r.status==St.GAP)
        score=round((done+partial*0.5)/len(self.reqs)*100)
        high_gaps=[r for r in self.reqs if r.status==St.GAP and "EU AI Act" in r.reg]
        verdict="READY" if gaps==0 and partial==0 else "CONDITIONAL" if not high_gaps else "NOT READY"

        print(f"COMPLIANCE REPORT: {self.proj}")
        print(f"Date: {datetime.now().strftime('%Y-%m-%d')} | Score: {score}% | Verdict: {verdict}")

        print(f"\n  Features ({len(self.feats)}):")
        for f in self.feats:
            print(f"    {'[X]' if f.ok else '[!]'} {f.name:<16} {f.tier:<10} {f.obs} obligations")

        print(f"\n  Requirements ({len(self.reqs)}):")
        ic={"done":"[X]","partial":"[~]","gap":"[!]"}
        for r in self.reqs:
            print(f"    {ic[r.status.value]} {r.id:>2}. {r.name:<20} {r.status.value:<8} {r.owner:<10} {r.due}")

        if gaps+partial>0:
            print(f"\n  Action items ({gaps+partial}):")
            for r in self.reqs:
                if r.status!=St.DONE:
                    urg="CRITICAL" if r.status==St.GAP else "MODERATE"
                    print(f"    [{urg}] #{r.id} {r.name} -> {r.owner} by {r.due}")

        print(f"\n  Timeline: Q2=close gaps | Q3=complete partials | Q4=re-audit | Aug 2026=enforcement")
        print(f"  Verdict: {verdict}")
        return {"score":score,"verdict":verdict}

r=CR("Netsetos GenAI Platform")
r.report()
```
