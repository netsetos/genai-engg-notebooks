```python
providers = [
    {'name': 'Deepgram Nova-3', 'cost_hr': 0.46, 'indian': 5},
    {'name': 'AssemblyAI Univ-2', 'cost_hr': 0.15, 'indian': 99},
    {'name': 'Google Chirp 3', 'cost_hr': 0.96, 'indian': 12},
    {'name': 'Azure Speech', 'cost_hr': 1.02, 'indian': 8},
    {'name': 'Groq Whisper Turbo', 'cost_hr': 0.04, 'indian': 99},
    {'name': 'Self-host whisper', 'cost_hr': 0.0, 'indian': 99},
]

monthly = 1000
hdr = f"{'Provider':<24} {'$/hr':<8} {'Monthly $':<12} {'Monthly INR':<14} {'Indian langs'}"
print('STT Provider Comparison (1000 hrs/mo):')
print(hdr)
print('=' * len(hdr))
for p in providers:
    cost_mo = p['cost_hr'] * monthly
    inr_mo = cost_mo * 84
    print(f"{p['name']:<24} ${p['cost_hr']:<7.2f} "
          f"${cost_mo:>8,.0f}    Rs {inr_mo:>8,.0f}    {p['indian']}")

cheapest = min(providers, key=lambda x: x['cost_hr'])
most = max(providers, key=lambda x: x['indian'])
print(f"\nCheapest: {cheapest['name']} (${cheapest['cost_hr']}/hr)")
print(f"Most Indian: {most['name']} ({most['indian']} langs)")
print("Best Indian accuracy: Sarvam Saaras V3 (code-switch)")
```

<details><summary>Solution</summary>


```python
configs = [
    {'name': 'Premium', 'vad': 10, 'stt': 200,
     'llm': 500, 'tts': 150, 'cpm': 0.15},
    {'name': 'Balanced', 'vad': 10, 'stt': 300,
     'llm': 400, 'tts': 200, 'cpm': 0.08},
    {'name': 'Budget', 'vad': 10, 'stt': 400,
     'llm': 300, 'tts': 100, 'cpm': 0.01},
    {'name': 'OAI Realtime', 'vad': 0, 'stt': 0,
     'llm': 350, 'tts': 0, 'cpm': 0.12},
]

print('Voice Agent Latency:')
print(f"{'Config':<15} {'VAD':<6} {'STT':<6} "
      f"{'LLM':<6} {'TTS':<6} {'Total ms'}")
print('=' * 52)
for c in configs:
    t = c['vad'] + c['stt'] + c['llm'] + c['tts']
    print(f"{c['name']:<15} {c['vad']:<6} "
          f"{c['stt']:<6} {c['llm']:<6} "
          f"{c['tts']:<6} {t}")

convos, avg = 10000, 3
print(f"\nMonthly cost ({convos:,} x {avg} min):")
for c in configs:
    usd = c['cpm'] * avg * convos
    inr = usd * 84
    print(f"  {c['name']:<15} ${usd:>7,.0f} (Rs {inr:>9,.0f})")

print("\nBudget vs Premium: 15x cost for ~50ms gain")
```
