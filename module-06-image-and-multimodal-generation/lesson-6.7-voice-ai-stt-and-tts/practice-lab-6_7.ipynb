```python
configs = [
    {'name': 'Premium', 'vad': 10, 'stt': 200,
     'llm': 500, 'tts': 150, 'cpm': 0.15},
    {'name': 'Balanced', 'vad': 10, 'stt': 300,
     'llm': 400, 'tts': 200, 'cpm': 0.08},
    {'name': 'Budget', 'vad': 10, 'stt': 400,
     'llm': 300, 'tts': 100, 'cpm': 0.01},
    {'name': 'OAI Realtime', 'vad': 0, 'stt': 0,
     'llm': 350, 'tts': 0, 'cpm': 0.12},
]

print('Voice Agent Latency:')
print(f"{'Config':<15} {'VAD':<6} {'STT':<6} "
      f"{'LLM':<6} {'TTS':<6} {'Total ms'}")
print('=' * 52)
for c in configs:
    t = c['vad'] + c['stt'] + c['llm'] + c['tts']
    print(f"{c['name']:<15} {c['vad']:<6} "
          f"{c['stt']:<6} {c['llm']:<6} "
          f"{c['tts']:<6} {t}")

convos, avg = 10000, 3
print(f"\nMonthly cost ({convos:,} x {avg} min):")
for c in configs:
    usd = c['cpm'] * avg * convos
    inr = usd * 84
    print(f"  {c['name']:<15} ${usd:>7,.0f} (Rs {inr:>9,.0f})")

print("\nBudget vs Premium: 15x cost for ~50ms gain")
```
