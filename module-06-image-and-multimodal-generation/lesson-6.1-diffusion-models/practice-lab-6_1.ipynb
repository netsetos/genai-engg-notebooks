## Exercise 1 (Easy): Visualize the Forward Process

Apply noise at 10 timesteps. Plot the destruction.

<details><summary>Solution</summary>


```python
def linear_schedule(T=1000, start=0.0001, end=0.02):
    betas = torch.linspace(start, end, T)
    alphas = 1.0 - betas
    return torch.cumprod(alphas, dim=0)

def add_noise(image, alpha_bar):
    noise = torch.randn_like(image)
    noisy = (torch.sqrt(alpha_bar) * image
             + torch.sqrt(1 - alpha_bar) * noise)
    return noisy

# Create sample image
img = torch.zeros(1, 64, 64)
img[0, 16:48, 16:48] = 1.0

alpha_bars = linear_schedule(1000)
steps = list(range(0, 1000, 100))

# Print key values
print('Forward Process - Alpha Bar Values:')
print(f'  t=0:   ab={alpha_bars[0]:.4f} (clean)')
print(f'  t=100: ab={alpha_bars[100]:.4f}')
print(f'  t=300: ab={alpha_bars[300]:.4f}')
print(f'  t=500: ab={alpha_bars[500]:.4f}')
print(f'  t=700: ab={alpha_bars[700]:.4f}')
print(f'  t=900: ab={alpha_bars[900]:.6f}')
print(f'  t=999: ab={alpha_bars[999]:.6f} (noise)')

# Verify: signal energy at each step
for t in [0, 250, 500, 750, 999]:
    ab = alpha_bars[t].item()
    signal_pct = ab * 100
    noise_pct = (1 - ab) * 100
    print(f'  t={t:4d}: signal={signal_pct:5.1f}%, '
          f'noise={noise_pct:5.1f}%')
```

## Exercise 2 (Medium): Compare Schedules

Linear vs cosine. Which preserves signal longer?


```python
# YOUR CODE
# TODO: cosine_schedule, compare at t=200,500,800
```

<details><summary>Solution</summary>


```python
def cosine_schedule(T=1000):
    steps = torch.arange(T + 1) / T
    ab = torch.cos((steps + 0.008) / 1.008
                   * torch.pi / 2) ** 2
    ab = ab / ab[0]
    betas = 1 - (ab[1:] / ab[:-1])
    betas = betas.clamp(max=0.999)
    alphas = 1.0 - betas
    return torch.cumprod(alphas, dim=0)

lin_ab = linear_schedule(1000)
cos_ab = cosine_schedule(1000)

print('Schedule Comparison (alpha_bar):')
print(f'{"t":>5} {"Linear":>10} {"Cosine":>10} '
      f'{"Winner":>10}')
print('-' * 40)
for t in [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 999]:
    l = lin_ab[t].item()
    c = cos_ab[t].item()
    winner = 'Cosine' if c > l else 'Linear'
    if abs(c - l) < 0.01:
        winner = 'Tie'
    print(f'{t:5d} {l:10.4f} {c:10.4f} {winner:>10}')

# Key insight
print('\nCosine preserves MORE signal in early steps')
print('(t=0-500) and destroys faster at the end.')
print('Linear destroys too aggressively in the middle.')

# Verify monotonicity
assert all(lin_ab[i] >= lin_ab[i+1]
           for i in range(len(lin_ab)-1)), 'Not monotonic!'
assert all(cos_ab[i] >= cos_ab[i+1]
           for i in range(len(cos_ab)-1)), 'Not monotonic!'
print('\nBoth schedules monotonically decreasing (verified)')
```
