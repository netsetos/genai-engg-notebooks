```python
# Pure Python: CSV creation + prompt building + metadata
import csv, json, os

products = [
    {'name': 'Silk Saree',
     'hindi': '\u0930\u0947\u0936\u092e\u0940 \u0938\u093e\u0921\u093c\u0940',
     'style': 'elegant silk saree on mannequin'},
    {'name': 'Chai Set',
     'hindi': '\u091a\u093e\u092f \u0938\u0947\u091f',
     'style': 'ceramic chai cups and kettle'},
    {'name': 'Jhumka Earrings',
     'hindi': '\u091d\u0941\u092e\u0915\u093e',
     'style': 'gold jhumka earrings, jewelry photo'},
    {'name': 'Kurta Set',
     'hindi': '\u0915\u0941\u0930\u094d\u0924\u093e \u0938\u0947\u091f',
     'style': 'cotton kurta pajama on hanger'},
    {'name': 'Brass Diya',
     'hindi': '\u092a\u0940\u0924\u0932 \u0926\u0940\u092f\u093e',
     'style': 'traditional brass oil lamp, diya'},
]

# Write CSV
os.makedirs('product_images', exist_ok=True)
with open('products.csv', 'w', newline='') as f:
    w = csv.DictWriter(f,
        fieldnames=['name', 'hindi', 'style'])
    w.writeheader()
    w.writerows(products)

print(f'CSV written: {len(products)} products')

# Build prompts
for i, prod in enumerate(products):
    prompt = (f"{prod['style']}, product photography, "
              f"white background, studio lighting, 8k")
    meta = {
        'prompt': prompt,
        'seed': i,
        'model': 'sd-2.1',
        'product': prod['name'],
        'hindi_name': prod['hindi'],
        'guidance_scale': 7.5,
        'steps': 25,
    }
    fname = f"product_images/{i:03d}_{prod['name'].replace(' ','_')}"
    with open(f'{fname}.json', 'w') as mf:
        json.dump(meta, mf, indent=2, ensure_ascii=False)
    print(f"  [{i+1}/{len(products)}] {prod['name']}"
          f" | {prod['hindi']}")
    print(f"    Prompt: {prompt[:60]}...")
    print(f"    Metadata: {fname}.json")

# Read back and verify
with open('products.csv') as f:
    reader = csv.DictReader(f)
    rows = list(reader)
print(f'\nCSV verified: {len(rows)} products')
for r in rows:
    print(f"  {r['name']} ({r['hindi']})")

# Count metadata files
meta_files = [f for f in os.listdir('product_images')
              if f.endswith('.json')]
print(f'\nMetadata files: {len(meta_files)}')
for mf in sorted(meta_files):
    with open(f'product_images/{mf}') as f:
        data = json.load(f)
    print(f"  {mf}: {data['product']} | seed={data['seed']}")
```
